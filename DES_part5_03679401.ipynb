* 5.3 Analysis and General Questions :

* Task 5.3.1: Confidence Interval Width :

In task 5.2.1, for the same simulation time we need more runs to reach the given width of the confidence interval if the confidence level is 95% campared to a confidence level of 90% which is normal since we are less confident for a confidence level of 90% that the true value is in our confidence interval so we need more runs to reach the given width of the confidence interval; otherwise we need more runs if the time simulation is 100s campared to 1000s because for a simulation of 1000s we are more confident that we can reach a steady state for the simulation.
In task 5.2.2, for the same confidence level we need more simulation time to reach the given width of the confidence interval if the number of packets N of one batch is 1000 campared to the number N of 100 and this can be explained by the fact that the simulation time for N of 1000 packets is longer than a simulation for N of 100 packets.

Using the batches is superior to replication of the same simulation because each replication contains a startup segment which may not be representative of steady behavior of the simulation; on the other hand the advantage of using the batches is that a single run can be used to obtain an estimate and only one transient state is included in the output.
