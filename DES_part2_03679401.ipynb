* 2.8 Analysis and General Questions :

* Task 2.8.1: Systems with Different Queue Capacity :

We observe from the plots that the mean waiting time distribution and the the mean queue length (buffer fill status) distribution are "Normal" distributions and that is predictable since the mean waiting time and the mean queue length are both a process that is the sum of a number of component processes (central limit theorem (CLT)).

We note from the plots that the mean of the mean waiting time distribution and also the mean of the mean queue length (buffer fill status) distribution is increasing when the queue size increases. We observe also that the variance of the distributions is increasing when the queue size increases.

We note that the variance of the distributions is decreasing since we have more stability when the simulation time increases so less variance for our system.  
For example we compute the mean and the variance of the mean queue length for a simulation time of 100s:

mean(S=5) = 2.54639254 ; Var(S=5) = 0.309241325682 ;
mean(S=6) = 3.04775935 ; Var(S=6) = 0.552278718659 ;
